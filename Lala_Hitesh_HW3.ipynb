## Problem 1: the coin flip

```python
res=np.random.choice([0,1], 8)
print("The coin flips are",res," with 0 = Heads, 1= Tails")
n=1

b = np.arange(0, 1,0.01 )
db = b[1] - b[0]

p_b=[[]]*8
fig, ax = plt.subplots(2,4, figsize=(22,11))
ax = ax.ravel()


for i in range(0,8):
    k=sum(j for j in res[0:i+1])

    p_b[i] = b ** k * (1 - b) ** (n - k)
    p_b[i]=np.array(p_b[i])
    p_b[i] /= p_b[i].sum()
    p_b[i] /= db
    
    


    ax[i].plot(b, p_b[i], '-k')
    
    if i>0:
        ax[i].plot(b, p_b[i-1], '--r')
        
        
    ax[i].set_xlabel('$b$', fontsize=15)
    ax[i].set_ylabel('$p(b|x,I)$', fontsize=15)
    ax[i].set_title("Black: Posterior with n="+str(n)+", k="+str(k)+"; Red:prior", fontsize=15)
    ax[i].set_ylim(-0.05, np.max(p_b[i])*1.1)
    n+=1
    
    
cuml_p_b = p_b[7].cumsum()
cuml_p_b /= cuml_p_b[-1]
x=cuml_p_b[b==0.25]
fig1= plt.figure(figsize=(10,8), )
axs=fig1.add_subplot(111, yscale='log')
axs.plot(b, cuml_p_b, '-m', label='Cumulative Distribution Function')

x=cuml_p_b[b==0.25]
fin=float(x[0])
axs.set_xlabel('$b$')
axs.set_ylabel('$P(<b|x,I)$')
axs.set_ylim(1E-6, 2)
axs.plot([0.25,0.25], [1E-6, 2], '-k')
axs.plot(0.25, x, "or", label='Desired Value= %.3f'%fin)
axs.legend()
print("The value for P(b<0.25) is: %.3f" %fin)
fin=float(x[0])
if fin >=0.01:
    print("We cannot reject the hypothesis that heads is preferred by our toss at 1% confidence level")
else :
    print("We can reject the hypothesis that heads is preferred by our toss at 1% confidence level")
    
print("\n\n\n\n")
```

## Problem 2: use BIC to find the best polynomial model for a given dataset

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() # set default plot styles
from scipy import stats
from scipy import optimize

# this function computes polynomial models given some data x
# and parameters theta
def polynomial_fit(theta, x):
    """Polynomial model of degree (len(theta) - 1)"""
    return sum(t * x ** n for (n, t) in enumerate(theta))

# compute the data log-likelihood given a model
def logL(theta, x, y, sigma_y, model=polynomial_fit):
    """Gaussian log-likelihood of the model at theta"""
    y_fit = model(theta, x)
    return sum(stats.norm.logpdf(*args)
               for args in zip(y, y_fit, sigma_y))

# a direct optimization approach is used to get best model 
# parameters (which minimize -logL)
def best_theta(degree, x, y, sigma_y, model=polynomial_fit):
    theta_0 = (degree + 1) * [0]
    neg_logL = lambda theta: -logL(theta, x, y, sigma_y, model)
    return optimize.fmin_bfgs(neg_logL, theta_0, disp=False)
```

```python
# generate (noisy) data
np.random.seed(0)
Ndata = 22
x = np.linspace(0, 3, Ndata)[1:-1]
sigma_y = 0.1 + 0*x
y = np.random.normal(np.sin(x) * x, sigma_y) 
data=np.stack((x,y, sigma_y))
xfit = np.linspace(-0.1, 3.1, 1000)

theta=[[] for _ in range(6)]
yfit =[[] for _ in range(6)]
chi2=np.zeros(6)
chi2dof=np.zeros(6)
bic=np.zeros(6)

fig, ax = plt.subplots(figsize=(16, 8))
ax.errorbar(x, y, sigma_y, fmt='ok', ecolor='gray')

name=["null","linear", "quadratic", "cubic", "4th degree", "5th degree"]

for i in range(1,6):
    
    theta[i] = best_theta(i, x, y, sigma_y)
    theta[i] = np.array(theta[i])
    yfit[i] = polynomial_fit(theta[i], xfit)
    yfit[i] = np.array(yfit[i])
    chi2[i] = np.sum(((y-polynomial_fit(theta[i], x))/sigma_y)**2) 
    chi2dof[i] = chi2[i]/(Ndata - i+1)
    bic[i]=chi2[i]+(i+1)*np.log(Ndata)
    ax.plot(xfit, polynomial_fit(theta[i], xfit), label="best "+str(name[i])+ " model")
    
    


ax.legend(loc='best', fontsize=14)
ax.set(xlabel='x', ylabel='y', title='data')

print("The reduced Chi2dof values are: ", chi2dof[1:])
print("The BIC values are: ", bic[1:])
```

As the BIC value of the cubic model is the smallest, it is the best model for the data among the 5 fitted models.

## Problem 3
Compare Bayesian Blocks Algorithm and classsical histogram
on a sample of periodic variable stars.

```python
# fetch data
data = fetch_LINEAR_geneva()
x = data['logP']
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, yscale='log')

hist(x, bins='knuth', ax=ax, density=True, histtype='stepfilled',color='c', alpha =0.9, label='Knuth Histogram')
hist(x, bins='blocks', ax=ax, density=True,histtype='step', color='k', label="Bayesian Blocks", lw=2.0)
hist(x,  bins =100, ax=ax, density=True, histtype='step', color='r', label="True Data", lw=1.0)
ax.set_xlim(-2.0, 2.0)
ax.set_ylabel('$p(x)$')
ax.legend(loc='upper right', prop=dict(size=8))
ax.set_xlabel('$x$')
ax.set_title("Log Space")
```
